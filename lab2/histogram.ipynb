```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img_path = "bay.jpg"
bay = cv2.imread(img_path)

plt.imshow(bay)
plt.axis('off')
```

```python
# thống kê histogram
hist = []
for i in range(256):
    hist.append(np.sum(bay == i))

plt.bar(range(256), hist)
plt.title("Histogram")

# vẽ bảng 
df = pd.DataFrame({'Pixel Value': range(256), 'Frequency': hist})
print(df)

```
